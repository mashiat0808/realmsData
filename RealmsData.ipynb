```python
# read all of the jsons in the directory, set the filename as a column
# and concat them all together
import glob
import pandas as pd
import json

jsons = glob.glob(os.path.join(REALMS_PROPOSALS_NODE_DATA_DIR, '*proposals.json'))

# read all the jsons into a flattened list
json_list = []
for json_file in jsons:
    with open(json_file, 'r') as f:
        read_json = json.load(f)
        # these files are a list of lists of dicts, so flatten them
        # if this is an empty list, skip
        if len(read_json) == 0:
            continue
        # if this is a list of dicts, flatten it
        if type(read_json[0]) == dict:
            read_json = [read_json]
        if type(read_json) != list:
            print(read_json)
        # for all objects in the list, set the filename as a key
        for obj in read_json:
            for i in obj:
                i['filename'] = os.path.basename(json_file)
                # replace _proposals.json at end with ''
                i['filename'] = i['filename'].replace('_proposals.json', '')
        json_list.extend(read_json)
```

```python
# remove empty lists from the json_list
json_list = [x for x in json_list if x != []]
# make json_list a 1d list instead of a list of lists
json_list = [item for sublist in json_list for item in sublist]
json_list
```

```python
import pandas as pd
import numpy as np

# Define size bins and labels
bins = [2, 10, 100, 1_000, 10_000, 100_000, 1_000_000, 10_000_000]
labels = ['[2,10]', '(10,100]', '(100,1K]', '(1K,10K]', '(10K,100K]', '(100K,1M]', '(1M,10M]']

# Filter DAOs with at least 2 unique voters
filtered_df = deployments1_df[deployments1_df["votes_count"] >= 2].copy()

# Assign DAO size category
filtered_df["size_category"] = pd.cut(filtered_df["votes_count"], bins=bins, labels=labels, right=True, include_lowest=True)

# Group by platform and size category
table = filtered_df.groupby(["size_category", "platform"]).size().unstack(fill_value=0)

# Add total per row (size category) and column (platform)
table["Total"] = table.sum(axis=1)
table.loc["Total"] = table.sum(axis=0)

# Reset index for display (optional)
table = table.reset_index()

# Expand display width to avoid line breaking
pd.set_option('display.width', 200)
pd.set_option('display.max_columns', None)

# Save or display
print(table)
```
